```python
import numpy as np
import pandas as pd
```

```python
x=np.array([[1,6],[1,8],[1,10],[1,14],[1,18]])
y=np.array([7,9,13,17.5,18])
X=x.T
A=np.matmul(X,x)
C=np.linalg.inv(A)
B=np.matmul(X,y)
print(np.matmul(C,B))
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
X=[[6],[8],[10],[14],[18]]
Q=np.linspace(5,20,100)
Y=[[7],[9],[13],[17.5],[18]]
reg=LinearRegression()
reg=reg.fit(X,Y)
y_pred=reg.predict(X)
print(y_pred)

yy = reg.predict(Q.reshape(Q.shape[0], 1))
plt.plot(Q,yy,c='r',linestyle='--')

r2=reg.score(X,Y)
print('r squared value is :%.6f'%r2)
```

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

d=sb.load_dataset('tips')
print(d)
```

```python
X=d['total_bill']
X=np.array(X)
A=X.reshape(244,1)
Y=d['tip']
y=np.array(Y)
B=y.reshape(-1,1)
plt.scatter(A,B)
plt.show()
```

```python
reg=LinearRegression()
reg=reg.fit(A,B)
y_pre=reg.predict(A)
plt.plot(A,y_pre)
r2=reg.score(A,B)
plt.scatter(A,B)
print(r2)
```

```python
x_mean=np.mean(A)
y_mean=np.mean(B)
num=0
den=0
n=len(A)
for i in range(n):
    num+=(A[i]-x_mean)*(B[i]-y_mean)
    den+=(A[i]-x_mean)**2
    
b1=num/den
b0=y_mean-(b1*x_mean)
print(b0,b1)
ssr=0
sst=0
C=b0+b1*A
plt.scatter(A,B)
plt.plot(A,C)

for i in range(n):
    y_pred=b0+(b1*A[i])
    sst+=(B[i]-y_mean)**2
    ssr+=(B[i]-y_pred)**2

r2=1-(ssr/sst)
print(r2)
```
