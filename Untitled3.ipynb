# Count Frequency

```python
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
```

```python
columns=[1,3,5,6,7,8,9,13]
data
```

```python
data=data[columns]
data.columns=['Employment','Degree','Relation','Designation','family','Race','Sex','Country']
```

```python
data
```

```python
for col in data.columns:
    print(col,len(data[col].unique()),'Values ')
    
```

```python
data['Country'].value_counts().to_dict()
```

```python
country_map=data['Country'].value_counts().to_dict()
```

```python
data['Country']=data['Country'].map(country_map)
```
