```python
import datetime
```

```python
df=datetime.datetime.today()
```

```python
df
```

```python
days=[df-datetime.timedelta(i) for i in range(1,15)]
```

```python
days
```

```python
ds=pd.DataFrame(days)
```

```python
ds.columns=['days']
```

```python
ds['wday']=ds['days'].dt.day_name()
```

```python
ds.head()
```

```python
week={
    'Monday':1,
    'Tuesday':2,
    'Wednesday':3,
    'Thursday':4,
    'Friday':5,
    'Saturday':6,
    'Sunday':7
}
```

```python
ds['day_rank']=ds['wday'].map(week)
```

```python
data=sb.load_dataset('titanic')
```

```python
data.head()
```

```python
col=['survived','deck']
data=data[col]
```

```python
s=data.groupby(['deck'])['survived'].mean().sort_values().index
```

```python
sa={k:i for i,k in enumerate(s,0)}
```

```python
sa
```

```python
data['deck']=data['deck'].map(sa)
```
