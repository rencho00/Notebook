```python
def miss(data,variable):
    mean_var=data[variable].mean()
    median_var=data[variable].median()
    mode_var=data[variable].mode()
    print('jdksajfkda',mean_var)
    print('sfdaf',median_var)
    print('sdada',mode_var)
```

```python
miss(data,'age')
```

```python
def normdist(data,variable):
    mean_var=data[variable].mean()
    median_var=data[variable].median()
    mode_var=data[variable].mode()
    
    plt.figure(figsize=(17,6))
    plt.subplot(1,3,1)
    data[variable].plot.kde()
    me=data[variable].fillna(mean_var)
    me.plot.kde()
    plt.subplot(1,3,2)
    data[variable].plot.kde()
    med=data[variable].fillna(median_var)
    med.plot(kind='kde')
    plt.subplot(1,3,3)
    data[variable].plot.kde()
    mod=data[variable].fillna(38)
    mod.plot(kind='kde')
    plt.legend()
    plt.show()
    
```

```python
normdist(data,'age')
```

```python
mod_var=data['age'].mode()
data['age']=data['age'].fillna(mod_var)
data.isnull().sum()
```

```python
def imp(df,variable):
    df[variable+'_random']=df[variable]
    ran_sam=df[variable].dropna().sample(df[variable].isnull().sum(),random_state=0)
    ran_sam.index=df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable+'_random']=ran_sam
    
```

```python
imp(df,'age')
```
